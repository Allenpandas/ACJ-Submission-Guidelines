```python
import pandas as pd
data = pd.read_csv('./discoveries.csv')
# print(data)
# 按照time排序
data.sort_values(by='time', inplace=True)
data.reset_index(drop=True, inplace=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = data[['value']]  # 取value的值

# plt.plot(year,data)
data.plot(subplots=True, figsize=(16, 9))


# 设置x轴和y周标题
plt.xlabel("year")
plt.ylabel("number of the discoveries")

plt.title('the number of the discoveries in the world')
plt.show()
```
